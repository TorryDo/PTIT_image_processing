```python
# import numpy as np
import matplotlib.pyplot as plt
import cv2

import pathlib
```

```python
dental_jpg = pathlib.Path('../../assets/hw1/dental.jpg')
parrot_jpg = pathlib.Path('../../assets/hw1/parrot.jpg')
skull_jpg = pathlib.Path('../../assets/hw1/skull.jpg')
```

```python
def adapt(img_path: str):
    
    img = cv2.imread(img_path, 0)

    hist_equalized = cv2.equalizeHist(img)

    clahe8x8 = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    img8x8 = clahe8x8.apply(img)

    clahe16x16 = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(16, 16))
    img16x16 = clahe16x16.apply(img)

    plt.figure(figsize=(12, 12))

    plt.subplot(2, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')

    plt.subplot(2, 2, 2)
    plt.imshow(hist_equalized, cmap='gray')
    plt.title('Global Histogram Equalization')

    plt.subplot(2, 2, 3)
    plt.imshow(img8x8, cmap='gray')
    plt.title('Adaptive Histogram Equalization, 8x8 tiles')

    plt.subplot(2, 2, 4)
    plt.imshow(img16x16, cmap='gray')
    plt.title('Adaptive Histogram Equalization, 16x16 tiles')
```

```python
adapt(str(parrot_jpg))
plt.show()
```

```python
adapt(str(skull_jpg))
plt.show()
```

```python
adapt(str(dental_jpg))
plt.show()
```
