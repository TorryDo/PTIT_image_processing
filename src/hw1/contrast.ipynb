```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv

# Contrast-limited histogram equalization
```

```python
img_path = '../../assets/hw1/moon.jpg'

original_img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)

clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
contrast_limit_hist = clahe.apply(original_img)

# show the original image
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.imshow(cv.cvtColor(original_img, cv.COLOR_BGR2RGB))
plt.title('Original image')

hist_original = cv.calcHist([original_img], [0], None, [256], [0, 256])
plt.subplot(2, 2, 3)
plt.bar(np.arange(256), hist_original.ravel())
plt.xlim([0, 255])
plt.xlabel('Gray level')
plt.ylabel('# pixels')
plt.title('Histogram of original image')

# show histogram of the original image

plt.subplot(2, 2, 2)
plt.imshow(cv.cvtColor(contrast_limit_hist, cv.COLOR_BGR2RGB))
plt.title('Contrast limit histogram equalization')

hist_original = cv.calcHist([contrast_limit_hist], [0], None, [256], [0, 256])
plt.subplot(2, 2, 4)
plt.bar(np.arange(256), hist_original.ravel())
plt.xlim([0, 255])
plt.xlabel('Gray level')
plt.ylabel('# pixels')
plt.title('Histogram of Contrast limit histogram equalization')

plt.show()
```
