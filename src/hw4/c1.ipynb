```python
M, N = 8, 8
U, V = 2.0, 2.0

COLS, ROWS = np.meshgrid(np.arange(M), np.arange(N))
print(COLS)
print('--------')
print(ROWS)
```

```python
I1 = 0.5 * np.exp(
    (1j * 2 * np.pi / 8) * 
    (U*COLS + V*ROWS)
)

DFT_I1 = np.fft.fftshift(np.fft.fft2(I1))

# show chart
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(np.real(I1), cmap='gray')
plt.title('REAL PART OF I1')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(np.imag(I1), cmap='gray')
plt.title('IMAGINARY PART OF I1')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(np.real(DFT_I1), cmap='gray')
plt.title('REAL PART OF DFT(I1)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(np.imag(DFT_I1), cmap='gray')
plt.title('IMAGINARY PART OF DFT(I1)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
