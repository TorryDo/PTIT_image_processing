```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M, N = 8, 8
U, V = 2.0, 2.0

COLS, ROWS = np.meshgrid(np.arange(M), np.arange(N))

I4 = np.sin(
    (2 * np.pi / 8) * 
    (U*COLS + V*ROWS)
)
DFT_I4 = np.fft.fftshift(np.fft.fft2(I4))


plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(I4, cmap='gray')
plt.title('I4')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(np.real(DFT_I4), cmap='gray')
plt.title('REAL PART OF DFT(I4)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(np.imag(DFT_I4), cmap='gray')
plt.title('IMAGINARY PART OF DFT(I4)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
