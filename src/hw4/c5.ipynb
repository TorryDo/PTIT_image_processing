```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import pathlib
```

```python
M, N = 8, 8
U, V = 1.5, 1.5

cols, rows = np.meshgrid(np.arange(M), np.arange(N))

I5 = np.cos(
    (2 * np.pi / 8) * 
    (U * cols + V * rows)
)


plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(I5, cmap='gray')
plt.title('I5')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(np.real(I5), cmap='gray')
plt.title('REAL PART OF I5')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(np.imag(I5), cmap='gray')
plt.title('IMAGINARY PART OF I5')
plt.axis('off')

plt.tight_layout()
plt.show()
```
