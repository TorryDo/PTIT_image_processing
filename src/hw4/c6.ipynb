```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import pathlib
```

```python
camera_data = np.fromfile(camerabin_path, dtype=np.uint8, count=256*256)
salesman_data = np.fromfile(salesman_path, dtype=np.uint8, count=256*256)
head_data = np.fromfile(head_path, dtype=np.uint8, count=256*256)
eyeR_data = np.fromfile(eyeR_path, dtype=np.uint8, count=256*256)

camera_data = camera_data.reshape(256, 256)

threshold_value = 128
binary_image = (camera_data > threshold_value).astype(np.uint8) * 255
```
